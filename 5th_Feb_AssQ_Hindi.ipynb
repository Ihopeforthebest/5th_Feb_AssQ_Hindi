```python
# Object-oriented programming (OOP) is a programming paradigm that allows you to model real-world entities as classes, 
# which are then used to create objects. A class is essentially a blueprint or a template that defines the properties and behaviors of an object, 
# while an object is an instance of a class that has specific values assigned to its properties.

# In simpler terms, a class is like a blueprint, and an object is like a building made using that blueprint.

# Let's take an example to understand this better.

# Consider the class Car. This class can have properties such as make, model, year, color, mileage, and behaviors 
# such as start, stop, accelerate, brake.

# class Car:
#     def __init__(self, make, model, year, color, mileage):
#         self.make = make
#         self.model = model
#         self.year = year
#         self.color = color
#         self.mileage = mileage
    
#     def start(self):
#         print("Car has started")
        
#     def stop(self):
#         print("Car has stopped")
    
#     def accelerate(self):
#         print("Car is accelerating")
        
#     def brake(self):
#         print("Car is braking")
# In the above code, we have defined the Car class with its properties and behaviors using methods. 
# Now, we can create multiple instances of this class, i.e., objects, each with its specific values assigned to the properties.

# car1 = Car("Toyota", "Corolla", 2020, "White", 5000)
# car2 = Car("Honda", "Civic", 2019, "Black", 10000)

# In the above code, we have created two objects, car1 and car2, using the Car class, 
# each with its specific values assigned to the properties. We can then access the methods of these objects to perform certain actions.

# car1.start()    # Output: Car has started
# car2.accelerate()   # Output: Car is accelerating

# In the above code, we have accessed the start method of car1 and the accelerate method of car2, which are specific to these objects.

# In summary, a class is a blueprint that defines the properties and behaviors of an object, 
# while an object is an instance of a class with specific values assigned to its properties.


```

```python
# Ans. 3)
```

```python
# In object-oriented programming, the __init__() function is a special method that is called when an object is created from a class. 
# It is also known as a constructor because it constructs or initializes the object by setting its initial state.

# The __init__() function is used to define and initialize the instance variables or attributes of an object. 
# These variables represent the state of the object and are unique to each instance of the class.

# Let's take an example to understand this better.

# class Person:
#     def __init__(self, name, age):
#         self.name = name
#         self.age = age
    
#     def introduce(self):
#         print(f"My name is {self.name} and I am {self.age} years old.")
# In the above code, we have defined a Person class with an __init__() method that takes two parameters, name and age. 
# These parameters are used to initialize the instance variables self.name and self.age, respectively.

# We have also defined a introduce() method, which will print out the name and age of the person.

# Now, we can create objects of the Person class and initialize their attributes using the __init__() method.

# person1 = Person("John", 30)
# person2 = Person("Jane", 25)
# In the above code, we have created two objects, person1 and person2, of the Person class and initialized their attributes
# using the __init__() method.

# We can then call the introduce() method on these objects to print out their names and ages.

# person1.introduce()  # Output: My name is John and I am 30 years old.
# person2.introduce()  # Output: My name is Jane and I am 25 years old.
# In summary, the __init__() function is used to initialize the attributes or instance variables of an object when it is created. 
# It allows us to set the initial state of the object and ensure that it has all the necessary information to perform its functions.
```

```python
# Ans. 4)
```

```python
# In object-oriented programming, self is a reference variable that refers to the current instance of a class. 
# It is a convention in Python to use the name self as the first parameter of instance methods in a class.

# When a method is called on an instance of a class, the instance itself is passed as the first argument to the method, 
# and this argument is conventionally named self. By using self, we can access the attributes and methods of the current instance of the class, 
# as well as modify its state.

# Here's an example to illustrate this:

# class Person:
#     def __init__(self, name, age):
#         self.name = name
#         self.age = age
    
#     def introduce(self):
#         print(f"My name is {self.name} and I am {self.age} years old.")
        
#     def change_name(self, new_name):
#         self.name = new_name
# In the above code, we have defined a Person class with an __init__() method that initializes the name and age attributes of the instance. 
# We have also defined two methods, introduce() and change_name(), which use the self parameter to access and modify the state of the instance.

# The introduce() method uses self to access the name and age attributes of the instance and prints them out. 
# The change_name() method takes a new_name parameter and uses self to set the name attribute of the instance to the new value.

# Now, we can create an instance of the Person class and call its methods using the self parameter.

# person1 = Person("John", 30)
# person1.introduce()  # Output: My name is John and I am 30 years old.
# person1.change_name("Jane")
# person1.introduce()  # Output: My name is Jane and I am 30 years old.
# In the above code, we have created an instance of the Person class and called its introduce() method using self. 
# We have also called the change_name() method to change the name attribute of the instance using self.

# In summary, self is used in object-oriented programming to refer to the current instance of a class. 
# It allows us to access and modify the state of the instance, as well as call its methods.
```

```python
# Inheritance is a concept in object-oriented programming that allows a class to inherit attributes and methods from another class. 
# The class that inherits is called the derived or child class, and the class that is being inherited from is called the base or parent class.

# There are four types of inheritance in Python:

# Single Inheritance
# Multiple Inheritance
# Multi-level Inheritance
# Hierarchical Inheritance
# Let's take an example for each type of inheritance:

# 1) Single Inheritance:
# In single inheritance, a derived class inherits from a single base class. This is the most common type of inheritance.

# python
# Copy code
# class Animal:
#     def __init__(self, name):
#         self.name = name

#     def eat(self):
#         print(f"{self.name} is eating.")

# class Dog(Animal):
#     def bark(self):
#         print(f"{self.name} is barking.")
        
# dog = Dog("Max")
# dog.eat()   # Output: Max is eating.
# dog.bark()  # Output: Max is barking.
# In the above example, we have a base class Animal with an eat() method. 
# We then define a derived class Dog that inherits from the Animal class and has an additional bark() method. 
# The Dog class can access the name attribute and eat() method of the Animal class using inheritance.

# 2) Multiple Inheritance:
# In multiple inheritance, a derived class inherits from multiple base classes.

# python
# Copy code
# class Flyable:
#     def fly(self):
#         print(f"{self.name} is flying.")

# class Walkable:
#     def walk(self):
#         print(f"{self.name} is walking.")

# class Bird(Flyable, Walkable):
#     def __init__(self, name):
#         self.name = name
        
# bird = Bird("Sparrow")
# bird.fly()  # Output: Sparrow is flying.
# bird.walk()  # Output: Sparrow is walking.
# In the above example, we have two base classes Flyable and Walkable with a fly() and walk() method, respectively. 
# We then define a derived class Bird that inherits from both the Flyable and Walkable classes. The Bird class can access 
# both the fly() and walk() methods using multiple inheritance.

# 3) Multi-level Inheritance:
# In multi-level inheritance, a derived class inherits from a base class, which in turn inherits from another base class.

# class Animal:
#     def __init__(self, name):
#         self.name = name

#     def eat(self):
#         print(f"{self.name} is eating.")

# class Dog(Animal):
#     def bark(self):
#         print(f"{self.name} is barking.")

# class Bulldog(Dog):
#     def __init__(self, name):
#         self.name = name
#         self.breed = "Bulldog"
        
# bulldog = Bulldog("Max")
# bulldog.eat()   # Output: Max is eating.
# bulldog.bark()  # Output: Max is barking.
# In the above example, we have a base class Animal with an eat() method. 
# We then define a derived class Dog that inherits from the Animal class and has an additional bark() method.
# We then define another derived class Bulldog that inherits from the Dog class and adds a breed attribute. 
# The Bulldog class can access the eat() and bark() methods of the Animal and Dog classes using multi-level inheritance.

# 4) Hierarchical Inheritance:
# If multiple derived classes are created from the same base, this kind of Inheritance is known as hierarchical inheritance. 
# In this instance, we have two base classes as a parent (base) class as well as two children (derived) classes.
# # Python program for demonstrating Hierarchical inheritance  
  
# # Here, we will create the Base class   
# class Parent1:  
#     def func_1(self):  
#         print ("This function is defined inside the parent class.")  
  
# # Derived class1  
# class Child_1(Parent1):  
#     def func_2(self):  
#         print ("This function is defined inside the child 1.")  
  
# # Derivied class2  
# class Child_2(Parent1):  
#     def func_3(self):  
#         print ("This function is defined inside the child 2.")  
  
# # Driver's code  
# object1 = Child_1()  
# object2 = Child_2()  
# object1.func_1()  
# object1.func_2()  
# object2.func_1()  
# object2.func_3() 

# Output:

# This function is defined inside the parent class.
# This function is defined inside the child 1.
# This function is defined inside the parent class.
# This function is defined inside the child 2.
```
